```python
import requests
# import json
import numpy as np
import matplotlib.pyplot as plt

def arbitrage(crypto, crncy='usd'):
    response = requests.get("https://api.cryptonator.com/api/full/{0}-{1}".format(crypto, crncy))
    live_api = response.json()

    bb = dict(live_api)
    ticker_var = dict(bb['ticker'])
    a = list(ticker_var['markets'])

    list1 = []

    for some_values in a:
        i = 0
        some_var = dict(some_values)
        list1.append(some_var['price'])
        i = i + 1

    list2 = []

    for some_more_values in a:
        c = 0
        some_more_val = dict(some_more_values)
        list2.append(some_more_val['market'])
        c = c + 1

    #MAX VALUE INDEX
    max_val = max(list1)
    max_val_index = list1.index(max_val)


    #MIN VALUE INDEX
    min_val = min(list1)
    min_val_index = list1.index(min_val)



    print('MAXIMUM PRICE', max_val, 'AT',  list2[max_val_index])
    print('MINIMUM PRICE', min_val, 'AT',  list2[min_val_index])
    profit = float(max_val)-float(min_val)
    print('WITH PROFIT OF', profit)

    numbers = np.arange(0, len(list1))

    plt.scatter(numbers, list1, color='red')

    for i, txt in enumerate(list2):
        plt.annotate(txt, (numbers[i], list1[i]))

    plt.title('ARBITRAGE')
    plt.ylabel('#PRICE')
    plt.show()
    
```

```python
arbitrage('btc')
```

```python
arbitrage('eth')
```
